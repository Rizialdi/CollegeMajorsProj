# Visualizing Earnings Based On College Majors
*We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).*

* Import the necessary libraries
* load the *magic command* **%matplotlib inline** for displaying plots
* Read the csv file and store in **recent_grads** variable
* Learn a little more about the data by printing the *.head()*, *.tail()*
* use Pandas.describes() function to give some statistical informations about the newly created dataFrame

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[:])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

### Remove from the dataset all null values

* null values can cause errors when plotting since it's undefined
* Used **Pandas.dropna()** function to remove rows containing a null value
* to see the actual process in action, counting the number of rows might be useful

```python
raw_data_count = recent_grads.count()
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.count()
print('Before dropping')
print(raw_data_count)
print('----------------------------')
print('After dropping')
print(cleaned_data_count)
```

* We moved from 173 to 172
* Just one row contained a null value

### Scatter plot to explore relation between *Sample_size* and *Median*

#### **Sample size : ** Sample size (unweighted) of full-time, year-round ONLY (used for earnings
#### **Median : **  Median earnings of full-time, year-round workers


* Scatter plot is an easy way to get quick insight about the *correlation* between two **randoms variables**
* DataFrame.plot() contains most of plotting functionalities in pandas

```python
recent_grads.plot(x='Sample_size', y='Median',
                  kind='scatter', title='Median vs. Sample_size', figsize=(8,6))
```

There are a lot of value of *Median* for the same *Sample_size*

**Median** and **Sample_size** are not correlated

### Scatter plot to explore relation between *Sample_size* and *Unemployment_rate*

#### *Unemployment_rate : * Unemployed / (Unemployed + Employed)

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate',
                  kind='scatter', title='Unemployment_rate vs. Sample_size', figsize=(8,6))
```

There are a lot of value of *Unemployment_rate* for the same *Sample_size*

**Unemployment** and **Sample_size** are not correlated

### Scatter plot to explore relation between *Full_time* and *Median*

#### *Full_time : * Employed 35 hours or more

```python
recent_grads.plot(x='Full_time', y='Median',
                  kind='scatter', title='Full_time vs. Median', figsize=(8,6))
```

There are a lot of value of *Full_time* for the same *Median*

**Full_time** and **Median** are not correlated

### Scatter plot to explore relation between *Unemployment_rate* and *ShareWomen*

#### ShareWomen : Women as share of total

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', 
                  kind='scatter', title='Unemployment_rate vs. ShareWomen', figsize=(8, 6))
```

### Scatter plot to explore relation between *Women* and *Median*

#### Women : Female graduates

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(8,6))
```

There are a lot of value of Full_time for the same Median

*Median* and *Women* are not correlated

### Use scatter_matrix from Pandas

* For seeing both Histogram and Scatter plot
* It helps to understand visually the distribution and the potential relationship simultaneously

```python
from pandas.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize = (10, 10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize = (10, 10))
```

```python
scatter_matrix(recent_grads[['Women', 'Unemployment_rate', 'Men']], figsize = (10, 10))
```

* => There's no relationship between the gender (Men and Women) and Unemployment rate for a major
